# Questão 1.c) Fatorial de n

```python
def fatorial(n):
  fat = 1
  if (n == 0): return 1
  for i in range(1,n+1):
    fat = fat * i
  return fat
```

```python
def fatorialrec(n):
  if (n == 0): return 1
  return n * fatorialrec(n-1)
```

# Questão 1.e) Manipulação de matrizes

```python
import random
```

```python
matriz = []

for i in range(0,5):
  linha = []
  for j in range(0,5):
    r = random.randrange(-10,10)
    linha  += [r]
  matriz += [linha]
```

```python
matriz
```

```python
soma = 0
media = 0
cont = 0
cont2 = 0
for i in range(0,5):
  for j in range(0,5):
    if matriz[i][j] < 0:
        cont = cont + 1
        matriz[i][j] = 0       
    if matriz[i][j] > 0:
        if (matriz[i][j] != 0):
            soma = soma + matriz[i][j]
            cont2 = cont2 + 1
    media = soma / cont2
```

```python
print(matriz)
print("Contagem de valores negativos = ", cont)
print("soma dos valores positivos = ",soma)
print("media dos valores positivos = ",media)
```

# Questão 1.f) Imagem em tons de cinza

```python
import numpy as np
from PIL import Image, ImageFont, ImageDraw
from PIL.ImageChops import add, subtract, multiply, difference, screen
import PIL.ImageStat as stat
from skimage.io import imread, imsave, imshow, show, imread_collection, imshow_collection
from skimage import color, viewer, exposure, img_as_float, data
from skimage.transform import SimilarityTransform, warp, swirl
from skimage.util import invert, random_noise, montage
import matplotlib.image as mpimage
import matplotlib.pylab as plt
from scipy.ndimage import affine_transform, zoom
from scipy import misc
```

```python
from PIL import Image

im = Image.open("lena_cor.bmp") # importa a imagem do diretorio:
map = im.load() # Cria a matriz de pixel:
  
# Extra a matriz
w, h = im.size 

for i in range(w):
    for j in range(h):   
        r, g, b = im.getpixel((i, j)) # Armazena o valor de pixel
          
        # Applica a formula:
        lena_cinza =  (r + g + b)/3 
  
        # setar um novo valor as cordenadas
        map[i, j] = (int(lena_cinza), int(lena_cinza), int(lena_cinza))

im.save("lena_cinza", format="png")
```

```python
im = Image.open("lena_cor.bmp") # importa a imagem do diretorio:
im2 = Image.open("lena_cinza")
fig, axs = plt.subplots(ncols=2, figsize=(10, 10))
axs[0].imshow(im, cmap=plt.cm.gray)
axs[0].set_title("Normal")
axs[1].imshow(im2, cmap=plt.cm.gray)
axs[1].set_title("escala de cinza")
```

formulas usadas:
converter um formato para outro
round((x / 255) * (2**bits - 1))

converter de volta
round(x / (2**bits - 1) * 255)

```python
im = Image.open("lena_cor.bmp") # importa a imagem do diretorio:
map = im.load() # Cria a matriz de pixel:
w, h = im.size

for i in range(w): 
    for j in range(h):
        r8, g8, b8 = im.getpixel((i, j)) 
        #Converte a imagem para 7bits
        sete_bits_r = round(((r8/255))* 127)
        sete_bits_g = round(((g8/255))* 127) 
        sete_bits_b = round(((b8/255))* 127)
      
        map[i, j] = (int(sete_bits_r),int(sete_bits_g),int(sete_bits_b))

        r, g, b = im.getpixel((i, j)) 
        #Converte de volta para 8bits
        sete_bits_r = round(r / (127) * 255)  
        sete_bits_g = round(g / (127) * 255)  
        sete_bits_b = round(b / (127) * 255)  
      
        map[i, j] = (int(sete_bits_r),int(sete_bits_g),int(sete_bits_b))
        im.save("lena_7con", format="bmp")
```

```python
im = Image.open("lena_cor.bmp") # importa a imagem do diretorio:
im2 = Image.open("lena_7con")
fig, axs = plt.subplots(ncols=2, figsize=(10, 10))
axs[0].imshow(im, cmap=plt.cm.gray)
axs[0].set_title("Normal 8bits")
axs[1].imshow(im2, cmap=plt.cm.gray)
axs[1].set_title("Quantizada em 7bits")
```
